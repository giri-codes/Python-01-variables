# PYTHON ESCAPE CHARACTERS:

Escape characters are special sequences in strings that start with a backslash \ and are used to represent characters that are hard or impossible to type directly (like new line, tab, quotes, etc.).

```python
# Example 1: New Line \n
text = "Hello\nWorld"
print(text)
```

```python
# Example 2: Tab Space \t
row = "Name\tAge\tCity"
data = "Giri\t25\tBangalore"

print(row)
print(data)
```

```python
# Example 3 (Medium): Double Quotes inside string
message = "He said \"Python is powerful\""
print(message)
```

```python
# Example 4 (Medium → Advanced): Backslash \\
path = "C:\\Users\\Giri\\Documents"
print(path)
```

```python
# Example 5 (Advanced): Carriage Return \r
text = "Hello World\rPython"
print(text)
```

## Real-Life Examples:

```python
# Example 1: Address Format
address = "Giri\nBangalore\nIndia"
print(address)
```

```python
# Example 2: Bill / Invoice Table
header = "Item\tPrice"
item1 = "Rice\t₹50"
item2 = "Milk\t₹30"

print(header)
print(item1)
print(item2)
```

```python
# Example 3: Chat Message
chat = "Friend said \"See you tomorrow\""
print(chat)
```

```python
# Example 4: File Location (Windows)
file_path = "D:\\Python\\Projects\\main.py"
print(file_path)
```

```python
# Example 5 (Advanced): Loading Status
status = "Loading...\rCompleted!"
print(status)
```
