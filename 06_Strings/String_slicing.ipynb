# String slicing:


```python
a = "I am a developer"    # First character = "index 0"
print(a[1:5])             # space is also counted.
```

```python
# To slice from start:

b = "I am a developer"
print(b[:12])  
```

```python
# Slice to the end:

c = "I am a developer"
print(c[0:]) 
```

```python
# Negative indexing:

d = "I am a developer"
print(d[-19:-1]) 
```

```python
# Examples:
# Example 1 (negative indexing)
e = "I am learning Python"
print(e[-20:-7])
```

```python
# Example 2 (positive to end)
f = "Python is powerful"
print(f[10:])
```

```python
# Example 3 (middle slice)
g = "Full Stack Developer"
print(g[5:16])
```

```python
# Example 4 (negative start, end omitted)
h = "Data Science Engineer"
print(h[-16:])
```

```python
# Example 5 (exclude last characters)

i = "BackendDeveloper"
print(i[:-5])
```
