# String Concatenation:

```python
# By using "+" operator:

a = "Gireesh"
b = "The Developer"
c = a + b
print(c)
```

```python
# To add a space add a " ":
a = "Gireesh"
b = "The Developer"
c = a + " " + b
print(c)
```

## Practice Examples:

```python
# Example 1: Join two words
a = "Hello"
b = "Universe!"

result = a + b
print(result)
```

```python
# Example 2: Add space between words
a = "Hello"
b = "Universe!"

result = a + " " + b
print(result)
```

```python
# Example 3: Sentence creation
name = "Giri"
msg = "Welcome"

print(msg + " " + name)
```

```python
# Example 4: Concatenate string with number (IMPORTANT)
# This gives error:
age = 25
print("Age is " + age)
```

```python
# Correct way:
age = 25
print("Age is " + str(age))
```

```python
# Example 5: Multiple concatenation
city = "Bangalore"
state = "Karnataka"
country = "India"

print(city + ", " + state + ", " + country)
```

## Real-Life String Concatenation Examples:

```python
# Example 1: Full Name (Forms / Certificates)
first_name = "Giri"
last_name = "Gowda"

full_name = first_name + " " + last_name
print(full_name)
```

```python
# Example 2: Phone Message
name = "Giri"
message = "Your OTP is "

otp = 1234
print(message + str(otp) + ", " + name)
```

```python
# Example 3: Shopping Bill Line
item = "Apple"
price = 50

print("Item: " + item + " | Price: ₹" + str(price))
```

```python
# Example 4: Student Result Message
student = "Giri"
result = "PASS"

print("Student " + student + " has " + result)
```

```python
# Example 5: Address Display (Delivery Apps)
house = "No 21"
area = "MG Road"
city = "Bangalore"

address = house + ", " + area + ", " + city
print(address)
```

```python
# Example 6: Login message
text = "Welcome!"
name = "Giri"
greeting = text + " " + name
print(greeting)
```

```python
# Example 7: Show Salary
text = "Salary"
amount = "50000"
show_salary = text + " ₹" + str(amount)
print(show_salary)
```

```python
# Example 8: Email format:
name = "giri"
format1 = "@gmail.com"
print("Your Email ID is " + name + format1)
```

```python
# Example 9: Product label with name + price
product = "Laptop"
price = 42000
product_label = product + " ₹" + str(price)

print(product_label)
```

```python
# Example 10: Date format 01-01-2026
day = 1
month = 1
year = 2026

date_format = str(day) + "-" + str( month) + "-" + str(year)
print(date_format)

# If you WANT 01-01-2026:

day = "01"
month = "01"
year = "2026"

date_format = day + "-" + month + "-" + year
print(date_format)
```
