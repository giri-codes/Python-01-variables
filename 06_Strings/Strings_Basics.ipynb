```python
# Strings :
a = "Hi G"
b = 'Hi G'
print(a)
print(b)
```

```python
# For quotes inside quotes:
a = "Hi 'Giri'!"   #for inside single quotes use outside double quotes.
b = "How are you!?"
c = 'I am "Fine"'  # for inside double quotes use outside single quotes.

print(a)
print(b)
print(c)
```

```python
# Multiline strings by using three double quotes:
a = """Hi i am Giri,
I am a effective learner and consistent doer,
I am a AI/ML learner
"""

print(a)
```

```python
# above example with three single quotes:
b = '''
Hi i am Giri,
I am a effective learner and consistent doer,
I am a AI/ML learner
'''

print(b)
```

```python
# String looping:
g = "Gireesh"
for g in "Gireesh":
    print(g)
```

```python
# To findout string length:
g = "Python Developer"
print(len(g))
```

```python
# To check string we use the keyword "in":

giri = "Python developer with 1 lakh salary per month!"
print("1 lakh" in giri)
```

```python
# We also check by using "if":

giri = "Python developer with 1 lakh salary per month!"
if "developer" in giri:
    print("Yes, 'developer' is present")
```

```python
# We use "not in" to check if a character is not present in a string:

giri = "Python developer with 1 lakh salary per month!"
print("5 lakhs" not in giri)
```

```python
# We can also check "not in" by using if:

giri = "Python developer with 1 lakh salary per month!"
if "boss" not in giri:
    print("No, 'boss' is not present")
```
