# Strings

## Definition
**String:**  
A string is a sequence of characters enclosed within single quotes (`' '`), double quotes (`" "`), or triple quotes (`""" """`) in Python.

## String — Syntax

```python
# string syntax
text = "Hello"
text = 'Hello'
text = """Hello"""
```

## String Slicing — Syntax

```python
# slicing syntax
# text[start:end]
# text[start:]
# text[:end]
# text[-start:-end]
```

## Modify Strings — Syntax

```python
# modify syntax
# text.upper()
# text.lower()
# text.strip()
# text.replace(old, new)
```

## String Concatenation — Syntax

```python
# concatenation syntax
# text1 + text2
```

## String Formatting — Syntax

```python
# formatting syntax
# f"{variable}"
# "{}".format(variable)
```

## Escape Characters — Syntax

```python
# escape characters
# "
# '
# \
# \n
# \t
```

## Real-Life Usage & Examples

### 1. User Name

```python
name = "Giri"
print("Welcome", name)
```

### 2. String Slicing

```python
text = "PythonDeveloper"
print(text[0:6])
```

### 3. Modify String

```python
text = "  python  "
print(text.strip().upper())
```

### 4. Concatenation

```python
first = "Python"
second = "Developer"
print(first + " " + second)
```

### 5. Formatting

```python
name = "Giri"
age = 25
print(f"My name is {name} and I am {age}")
```

## **MOST IMPORTANT INTERVIEW QUESTIONS & ANSWERS**

**Q1. What is a string in Python?**  
A string is a sequence of characters enclosed in quotes.

**Q2. Are strings mutable?**  
No, strings are immutable.

**Q3. What is string slicing?**  
Extracting a portion of a string using indexes.

**Q4. How do you concatenate strings?**  
Using the + operator.

**Q5. How do you format strings?**  
Using f-strings or format().

**Q6. What are escape characters?**  
Special characters used inside strings.

**Q7. What does strip() do?**  
Removes leading and trailing spaces.

**Q8. Difference between upper() and lower()?**  
upper() converts to uppercase, lower() converts to lowercase.

**Q9. What happens if you change a string?**  
A new string is created.

**Q10. Where are strings used?**  
In web apps, automation, data processing, and testing.
