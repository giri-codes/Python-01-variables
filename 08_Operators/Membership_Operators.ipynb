# MEMBERSHIP OPERATORS:

Membership operators are used to check whether a value exists inside a sequence (like a list, tuple, string, set, or dictionary).

| Operator | Meaning                                  |
| -------- | ---------------------------------------- |
| `in`     | Value **exists** in the sequence         |
| `not in` | Value **does NOT exist** in the sequence |

**Syntax**:

value `in` sequence

value `not in` sequence

---

`in` → checks presence → True if found

`not in` → checks absence → True if not found


```python
# List:
animals = ["Lion", "tiger", "wolf", "elephant"]

print("wolf" in animals)
```

```python
numbers = [10, 20, 30, 40]

print(50 in numbers)
```

```python
# Strings:
name = "Giri"
print("r" in name)
```

```python
goal = "developer"
print("g" not in goal)
```

```python
# Tuple:
colors = ("red", "green", "blue")

print("green" in colors)
```

```python
# Set:
vowels = {"a", "e", "i", "o", "u"}

print("e" in vowels)
```

```python
# Dictionary:
student = {"name": "Giri", "age": 25}

print("name" in student)

# Dictionary checks KEYS only, not values
```

```python
student = {"name": "Giri", "age": 25}

print("Giri" in student)
```

```python
# Example 1: Username Login Check
users = ["admin", "giri", "user1"]
username = "giri"

result = username in users
print(result)

# Output:
# True
```

```python
# Example 2: Attendance Verification
present_students = ["Ravi", "Giri", "Anu"]
student = "Kiran"

result = student in present_students
print(result)

# Output:
# False
```

```python
# Example 3: Product Availability
products = ["laptop", "mobile", "charger"]
product_name = "tablet"

result = product_name not in products
print(result)

# Output:
# True
```

```python
# Example 4: Email Format Check
email = "giri@gmail.com"
required_symbol = "@"

result = required_symbol in email
print(result)

# Output:
# True
```

```python
# Example 5: Website Access Restriction
blocked_sites = ["facebook.com", "instagram.com"]
current_site = "youtube.com"

result = current_site not in blocked_sites
print(result)

# Output:
# True
```

## REAL WORLD USAGES:

- **Login Validation** – Check if a username exists in allowed users list  
- **Access Control** – Verify user roles or permissions  
- **Input Validation** – Ensure required characters or words are present  
- **Data Filtering** – Include or exclude items from collections  
- **Security Checks** – Block restricted resources or content


## INTERVIEW Q & A:
1. **What are membership operators in Python?**  

Membership operators are used to test whether a value exists or does not exist in a sequence and return True or False.

2. **Which membership operators are available in Python?**  

Python provides two membership operators: `in` and `not in`.

3. **How does the `in` operator work with dictionaries?**  

It checks only the keys of the dictionary, not the values.

4. **What will be the output of `"p" in "python"`?**  

True, because the character "p" exists in the string.

5. **Can membership operators be used with integers directly?**  

No, because integers are not iterable; membership operators work only with iterable objects.
