# Logical Operators:
Logical operators are used to combine or modify conditions and return True or False based on logic.

| Operator | Name        | Meaning (Simple)                    |
| -------- | ----------- | ----------------------------------- |
| `and`    | Logical AND | All conditions must be True         |
| `or`     | Logical OR  | At least one condition must be True |
| `not`    | Logical NOT | Reverses the result                 |


```python
# and operator:
age = 27
id_no = True

print(age > 18 and id_no)
```

```python
# or operator:
salary = 15000
bonus = 5000

print(salary > 20000 or bonus > 3000)
```

```python
# not operator:
log_in = False
print(not log_in)
```

```python
# Logical Operators with Comparisons:
marks = 80
print(marks >= 35 and marks <= 100)
```

```python
# Scenario: Checking exam pass eligibility
# Condition: Student must score >= 35 in theory AND practical
theory_marks = 40
practical_marks = 38

result = theory_marks >= 35 and practical_marks >= 35
print(result)

# EDGE CASE: If any one subject < 35 → Fail
# Output: True
```

```python
# Scenario: Applying for driving license
# Condition: Age must be >= 18 AND learner license must be present
age = 18
has_learner_license = True

eligible = age >= 18 and has_learner_license
print(eligible)

# EDGE CASE: Minimum legal age
# Output: True
```

```python
# Scenario: Festival discount
# Condition: Purchase amount > 5000 OR user is premium member
purchase_amount = 5000
is_premium_member = False

discount = purchase_amount > 5000 or is_premium_member
print(discount)

# EDGE CASE: Amount exactly at limit but not premium
# Output: False
```

```python
# Scenario: Exam result calculation
# Condition SHOULD BE: Both subjects >= 35
theory_marks = 40
practical_marks = 20

result = theory_marks >= 35 or practical_marks >= 35
print(result)

# TRAP CASE: Using OR instead of AND
# Output: True (WRONG — student failed one subject)
```

```python
# Scenario: Pass validation
# Condition SHOULD BE: marks >= 35
theory_marks = 35
practical_marks = 36

result = theory_marks > 35 and practical_marks > 35
print(result)

# TRAP CASE: Using > instead of >=
# Output: False (WRONG — exact boundary treated as fail)
```

## INTERVIEW Q&A

1. **What are logical operators in Python?**  
   Logical operators are used to combine, evaluate, or reverse Boolean conditions and return `True` or `False`.

2. **What is the difference between `and` and `or`?**  
   `and` requires all conditions to be True, while `or` requires at least one condition to be True.

3. **What does the `not` operator do?**  
   The `not` operator reverses the Boolean value of a condition.

4. **What type of values do logical operators return?**  
   Logical operators always return Boolean values.

5. **Where are logical operators commonly used in real-world applications?**  
   They are commonly used in authentication, validation, access control, and decision-making logic.
