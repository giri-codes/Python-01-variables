# PYTHON OPERATORS:

**Operators are used to perform operations on variables and values**

## Arithmetic Operators:
#### **A S M D M E F** = 

| Operator | Name                | Example  |
| -------- | ------------------- | -------- |
| `+`      | Addition            | `a + b`  |
| `-`      | Subtraction         | `a - b`  |
| `*`      | Multiplication      | `a * b`  |
| `/`      | Division            | `a / b`  |
| `%`      | Modulus (remainder) | `a % b`  |
| `**`     | Exponent (power)    | `a ** b` |
| `//`     | Floor Division      | `a // b` |


## BODMAS Table:

| 1Ô∏è‚É£      | **B**rackets       | Solve inside `()` first | `( )`        |

```python
# Addition:
a = 10
b = 5
print(a + b)
```

```python
# Subtraction:
c = 20
d = 8
print(c - d)
```

```python
# Multiplication:
d = 6
e = 4
print(d * e)
```

```python
# Division (Always returns a float):
f = 10
g = 2
print(f / g)
```

```python
# Modulus (Remainder):
h = 10
i = 3
print(h % i)
```

```python
# Floor Division (No decimal):
j = 15
k = 5
print(j // k)

# Floor division always returns an integer.

# It rounds DOWN to the nearest integer
```

```python
# Exponentiation : Power Operator:
l = 2
m = 4
print(l ** m)
```

```python
# Mixed Operators:
n = 10
o = 2
p = 3

print(n + o * p) #(* runs first)
```

```python
# Using Variables:
q = 5
r = 5
print((q + r) ** 2)
```

```python
s = 20
t = 3
result = (s // t) + (s % t) * (t ** 2)
print(result)
```
